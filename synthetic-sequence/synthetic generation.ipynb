```python
import numpy as np
np.random.seed(0)

import random
random.seed(0)
```

```python
MAX_LEN = 20
MAX_EACH = MAX_LEN // 2
NUM_DATA = 5000
```

```python
def make_str(existing, a='a', b='b'):
    randstr = None
    while randstr is None or randstr in existing:
        num = random.randrange(MAX_EACH)
        letters = list(a * num + b * num)
        random.shuffle(letters)
        randstr = ''.join(letters)
    return randstr
```

```python
existing = set()
for _ in range(NUM_DATA):
    existing.add(make_str(existing))
```

```python
data = []
```

```python
for y in existing:
    data.append((np.random.rand(4), y))
```
